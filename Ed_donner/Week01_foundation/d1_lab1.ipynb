```python
from dotenv import load_dotenv
import os
load_dotenv()

try:
    api_key = os.getenv('GROQ_API_KEY')
    model = os.getenv('GROQ_REASONING_MODEL')
    print(model)
    if api_key:
        print("API key found successfully")
except Exception as e:
    print(e)
```

```python
from groq import Groq

client = Groq()
message = [
    {
        'role': 'user',
        'content': 'what is 2+2?'
    },
]

response = client.chat.completions.create(
    model=model,
    messages=message
)

print(response.choices[0].message.content)
```

```python
# creating question for accessing the iq of someone
message = [
    {
        'role': 'user',
        'content': 'please provide/propose a hard, challenging question to access someones IQ. Respond only with question'
    }
]

response = client.chat.completions.create(
    model=model,
    messages=message
)
question = response.choices[0].message.content
print(question)
```

```python
# generating ans from the LLM generated question
message = [
    {
        'role': 'user',
        'content': question
    }
]

response = client.chat.completions.create(
    model=model,
    messages=message
)
answer=response.choices[0].message.content
print(answer)
```
